```python
import pixelprism.math as pm
import pixelprism.math.functional as F
import numpy as np
```

```python
matrix_A = pm.var("A", pm.DType.FLOAT32, shape=(2, 2))
reshape_expr = F.reshape(matrix_A, (4,))
```

```python
with pm.new_context():
    pm.set_value("A", np.array([[1, 2], [3, 4]]))
    print(pm.get_value("A"))
    print(reshape_expr.eval())
# end with
```

```python
matrix_B = pm.var("B", pm.DType.FLOAT32, shape=(2, 2, 2))
reshape_expr = F.reshape(matrix_B, (-1, 2))
```

```python
with pm.new_context():
    pm.set_value("B", np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
    print(pm.get_value("B"))
    print(reshape_expr.eval().shape)
# end with
```

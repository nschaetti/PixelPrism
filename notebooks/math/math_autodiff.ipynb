```python
import pixelprism.math as pm
import pixelprism.math.functional as F
from pixelprism.math.render import render_latex
```

```python
x = pm.var("x", pm.DType.FLOAT32, pm.Shape.scalar())
```

```python
dx_dx = x.diff(wrt=x)
dx_dx.eval()
```

```python
c = pm.const("c", data=10, dtype=pm.DType.FLOAT32)
dc_dx = c.diff(wrt=x)
dc_dx.eval()
```

```python
y = x + c
dy_dx = y.diff(wrt=x)
dy_dx
```

```python
render_latex(y)
```

```python
render_latex(dy_dx)
```

```python
y = x * x
dy_dx = y.diff(wrt=x)
dy_dx
```

```python
render_latex(dy_dx)
```

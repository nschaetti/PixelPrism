# Math Statistics and Sampling
This notebook demonstrates sampling builders and statistics operators.

```python
import numpy as np
import pixelprism.math as pm
import pixelprism.math.functional.stats as S
```

```python
x = pm.var("nb_stats_x", dtype=pm.DType.R, shape=(5, 2))
ddof = pm.var("nb_stats_ddof", dtype=pm.DType.Z, shape=())
loc = pm.var("nb_stats_loc", dtype=pm.DType.R, shape=())
scale = pm.var("nb_stats_scale", dtype=pm.DType.R, shape=())

cov_expr = S.cov(x, rowvar=False, ddof=ddof)
corr_expr = S.corr(x, rowvar=False)
z_expr = S.zscore(x, axis=0, ddof=ddof)
normal_expr = S.normal(shape=(3, 3), loc=loc, scale=scale)
```

```python
with pm.new_context():
    pm.set_value("nb_stats_x", np.array([[1.0,2.0],[2.0,3.0],[3.0,6.0],[4.0,8.0],[5.0,9.0]], dtype=np.float32))
    pm.set_value("nb_stats_ddof", 1)
    pm.set_value("nb_stats_loc", 2.0)
    pm.set_value("nb_stats_scale", 0.5)
    print("cov:
", cov_expr.eval().value)
    print("corr:
", corr_expr.eval().value)
    print("zscore:
", z_expr.eval().value)
    print("normal:
", normal_expr.eval().value)
```

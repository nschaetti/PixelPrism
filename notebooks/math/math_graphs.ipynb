# Math Graphs
This notebook demonstrates graph primitives, graph operators, and expression graph conversion.

```python
import numpy as np
import pixelprism.math as pm
import pixelprism.math.functional as F
import pixelprism.math.functional.graph as G
```

```python
g = pm.digraph(n_nodes=4, edges=[(0,1),(1,2),(0,3)], allow_cycles=False)
print("is_cyclic:", g.is_cyclic())
print("topological_sort:", g.topological_sort())

adj_expr = pm.const("nb_adj", data=g.adjacency(), dtype=pm.DType.Z)
print("out_degree:", G.out_degree(adj_expr).eval().value)
print("laplacian:
", G.laplacian(adj_expr, directed=True).eval().value)
```

```python
x = pm.var("nb_graph_x", dtype=pm.DType.R, shape=())
y = pm.var("nb_graph_y", dtype=pm.DType.R, shape=())
expr = F.mul(F.add(x, y), y)
view = pm.expression_to_graph(expr)
print("nodes, edges:", view.graph.n_nodes, view.graph.n_edges)
print("adjacency:
", pm.expression_to_adjacency(expr).eval().value)
print(pm.graph_to_dot(view.graph))
```

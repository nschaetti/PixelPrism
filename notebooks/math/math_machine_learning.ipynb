# Math Machine Learning
This notebook demonstrates Perceptron, Decision Tree, and linear SVC (requires scikit-learn).

```python
import numpy as np
import pixelprism.math as pm
```

```python
try:
    x_bin = pm.var("nb_ml_x_bin", dtype=pm.DType.R, shape=(4, 2))
    y_bin = pm.var("nb_ml_y_bin", dtype=pm.DType.Z, shape=(4,))

    perceptron = pm.Perceptron(max_iter=200, learning_rate=1.0, shuffle=False)
    perceptron.fit(x_bin, y_bin)

    svc = pm.SVC(c=1.0, max_iter=2000)
    svc.fit(x_bin, y_bin)

    x_mul = pm.var("nb_ml_x_mul", dtype=pm.DType.R, shape=(6, 2))
    y_mul = pm.var("nb_ml_y_mul", dtype=pm.DType.Z, shape=(6,))
    tree = pm.DecisionTreeClassifier(max_depth=3, criterion="gini")
    tree.fit(x_mul, y_mul)

    with pm.new_context():
        pm.set_value("nb_ml_x_bin", np.array([[2.0,2.0],[1.0,1.0],[-1.0,-1.0],[-2.0,-2.0]], dtype=np.float32))
        pm.set_value("nb_ml_y_bin", np.array([1,1,-1,-1], dtype=np.int64))
        pm.set_value("nb_ml_x_mul", np.array([[-2.0,-2.0],[-1.0,-1.0],[0.0,0.0],[0.5,0.5],[2.0,2.0],[2.5,2.5]], dtype=np.float32))
        pm.set_value("nb_ml_y_mul", np.array([0,0,1,1,2,2], dtype=np.int64))

        print("Perceptron predict:", perceptron.predict(x_bin).eval().value)
        print("Perceptron theta:", perceptron.theta_.eval().value)
        print("Tree predict:", tree.predict(x_mul).eval().value)
        print("Tree classes:", tree.classes_.eval().value)
        print("SVC scores:", svc.decision_function(x_bin).eval().value)
        print("SVC predict:", svc.predict(x_bin).eval().value)
except ImportError as exc:
    print("Install scikit-learn to run ML examples:", exc)
```

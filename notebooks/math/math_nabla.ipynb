```python
import pixelprism.math as pm
import pixelprism.math.functional as F
from pixelprism.math.render import render_latex
```

```python
x = pm.var("x", pm.DType.FLOAT32, pm.Shape.scalar())
z = pm.var("z", pm.DType.FLOAT32, pm.Shape.scalar())
c = pm.const("c", 2.0)
y = x * z - c
```

```python
render_latex(y)
```

```python
dy_dx = F.diff(y, x)
```

```python
dy_dz = F.diff(y, z)
```

```python
render_latex(dy_dx)
```

```python
ny = F.nabla(y)
```

```python
render_latex(ny)
```

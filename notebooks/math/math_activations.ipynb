```python
import pixelprism.math as pm
import pixelprism.math.functional as F
from pixelprism.math import DType

matrix_A = pm.const("A", data=[[1.5, 2.5],[-1.0, 3.5]], dtype=pm.DType.FLOAT32)
matrix_B = pm.const("B", data=[[1.0, 0.0],[0.0, 1.0]], dtype=pm.DType.FLOAT64)
matrix_C = pm.const("C", data=[[0.5, 3.6],[1.2, 2.2]], dtype=pm.DType.INT32)
matrix_D = pm.var("D", dtype=pm.DType.FLOAT32, shape=pm.Shape.matrix(2, 2))

def inspect(expr):
    tensor = expr.eval()
    print(f"{expr.name}: shape={tensor.shape.dims}, dtype={tensor.dtype}")
    print(tensor.value)
    return tensor
# end def inspect
```

```python
sig_expr = F.sigmoid(matrix_A + matrix_D)
with pm.new_context():
    pm.set_value(matrix_D.name, [[1.0, 1.0],[1.0, 1.0]])
    inspect(sig_expr)
# end with
```

```python
relu_expr = F.relu(matrix_A + matrix_D)
with pm.new_context():
    pm.set_value(matrix_D.name, [[-10.0, 1.0],[1.0, 1.0]])
    inspect(relu_expr)
# end with
```

```python
gelu_expr = F.gelu(matrix_A + matrix_D)
with pm.new_context():
    pm.set_value(matrix_D.name, [[-10.0, 1.0],[1.0, 1.0]])
    inspect(gelu_expr)
# end with
```

```python
softplus_expr = F.softplus(matrix_A + matrix_D)
with pm.new_context():
    pm.set_value(matrix_D.name, [[-10.0, 1.0],[1.0, 1.0]])
    inspect(softplus_expr)
# end with
```

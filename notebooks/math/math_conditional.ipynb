# Scalar Conditional Operators

This notebook demonstrates the `if_` helper exposing the scalar `IfOperator`.

```python
import pixelprism.math as pm
import pixelprism.math.functional as F
import numpy as np
```

```python
x = pm.const("x", data=3.0, dtype=pm.DType.FLOAT32)
y = pm.const("y", data=-1.0, dtype=pm.DType.FLOAT32)
cond = pm.const("cond", data=True, dtype=pm.DType.BOOL)
expr = F.if_(cond, x, y)
print(expr.eval().value)  # -> 3.0
```

```python
# Demonstrate that the false branch is not evaluated
cond_false = pm.const("cond_false", data=False, dtype=pm.DType.BOOL)
unbound_branch = pm.var("unbound", dtype=pm.DType.FLOAT32, shape=())
safe_expr = F.if_(cond_false, unbound_branch, y)
print(safe_expr.eval().value)  # -> -1.0
print(safe_expr.shape.dims)
print(safe_expr.dtype)
```

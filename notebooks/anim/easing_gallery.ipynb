# PixelPrism Easing Gallery

Visualisation des fonctions de easing exposees dans `pixelprism.anim.easing`.

```python
import numpy as np
import matplotlib.pyplot as plt

from pixelprism.anim.easing import EASINGS, INTERPOLATES, create_interpolate

u = np.linspace(0.0, 1.0, 500)
easing_names = sorted(EASINGS.keys())
interpolate_names = sorted(INTERPOLATES.keys())

start_value = 10
end_value = 100

n = len(easing_names)
cols = 3
rows = int(np.ceil(n / cols))

fig, axes = plt.subplots(rows, cols, figsize=(15, 4 * rows), squeeze=False)

for idx, name in enumerate(easing_names):
    r = idx // cols
    c = idx % cols
    ax = axes[r][c]

    fn = create_interpolate(
        start=start_value,
        end=end_value,
        interpolate_fn=INTERPOLATES["integer"],
        easing_fn=EASINGS[name],
    )

    y = np.array([fn(t=float(x)) for x in u])
    ax.plot(u, y, linewidth=2, label=name)
    ax.plot([0, 1], [0, 1], linestyle='--', linewidth=1, color='gray', alpha=0.6, label='linear ref')
    ax.set_title(name)
    ax.set_xlim(0, 1)
    ax.set_ylim(min(-0.3, y.min() - 0.05), max(1.3, y.max() + 0.05))
    ax.grid(alpha=0.25)

for idx in range(n, rows * cols):
    r = idx // cols
    c = idx % cols
    axes[r][c].axis('off')

plt.tight_layout()
plt.show()
```

## Parametres de `ease_in_back` et `ease_*_elastic`

Exemples concrets pour ajuster `overshoot` et les parametres elastiques.

```python
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(0.0, 1.0, 600)

fig, axes = plt.subplots(2, 2, figsize=(14, 10), squeeze=False)

# --- ease_in_back avec overshoot variable
ax = axes[0][0]
for overshoot in [0.05, 1.0, 1.70158, 2.5]:
    fn = create_interpolate(
        start=start_value,
        end=end_value,
        interpolate_fn=INTERPOLATES["interpolate"],
        easing_fn=EASINGS["ease_in_back"],
        overshoot=overshoot
    )
    y = np.array([fn(t=float(x)) for x in u])
    ax.plot(u, y, linewidth=2, label=f'overshoot={overshoot}')
# end for
ax.plot([0, 1], [0, 1], '--', color='gray', alpha=0.6)
ax.set_title('make_ease_in_back(overshoot=...)')
ax.set_xlim(0, 1)
ax.grid(alpha=0.25)
ax.legend()

# --- ease_in_elastic: amplitude et period
ax = axes[0][1]
elastic_in_configs = [(0.7, 0.25), (1.0, 0.30), (1.3, 0.35)]
for amplitude, period in elastic_in_configs:
    fn = create_interpolate(
        start=start_value,
        end=end_value,
        interpolate_fn=INTERPOLATES["interpolate"],
        easing_fn=EASINGS["ease_in_elastic"],
        amplitude=amplitude,
        period=period
    )
    y = np.array([fn(t=float(x)) for x in u])
    ax.plot(u, y, linewidth=2, label=f'overshoot={overshoot}')
# end for
ax.plot([0, 1], [0, 1], '--', color='gray', alpha=0.6)
ax.set_title('make_ease_in_back(overshoot=...)')
ax.set_xlim(0, 1)
ax.grid(alpha=0.25)
ax.legend()

# --- ease_in_elastic: amplitude et period
# ax = axes[0][1]
# elastic_in_configs = [(0.7, 0.25), (1.0, 0.30), (1.3, 0.35)]
# for amplitude, period in elastic_in_configs:
#     fn = make_ease_in_elastic(amplitude=amplitude, period=period)
#     y = np.array([fn(float(x)) for x in u])
#     ax.plot(u, y, linewidth=2, label=f'amp={amplitude}, period={period}')
# ax.plot([0, 1], [0, 1], '--', color='gray', alpha=0.6)
# ax.set_title('make_ease_in_elastic(amplitude, period)')
# ax.set_xlim(0, 1)
# ax.grid(alpha=0.25)
# ax.legend()
#
# # --- ease_out_elastic: amplitude et period
ax = axes[1][0]
elastic_out_configs = [(0.7, 0.25), (1.0, 0.30), (1.3, 0.35)]
for amplitude, period in elastic_out_configs:
    fn = create_interpolate(
        start=start_value,
        end=end_value,
        interpolate_fn=INTERPOLATES["interpolate"],
        easing_fn=EASINGS["ease_out_elastic"],
        amplitude=amplitude,
        period=period
    )
    y = np.array([fn(float(x)) for x in u])
    ax.plot(u, y, linewidth=2, label=f'amp={amplitude}, period={period}')
ax.plot([0, 1], [0, 1], '--', color='gray', alpha=0.6)
ax.set_title('make_ease_out_elastic(amplitude, period)')
ax.set_xlim(0, 1)
ax.grid(alpha=0.25)
ax.legend()

# --- ease_in_out_elastic: amplitude et period
ax = axes[1][1]
elastic_in_out_configs = [(0.7, 0.40), (1.0, 0.45), (1.3, 0.55)]
for amplitude, period in elastic_in_out_configs:
    fn = create_interpolate(
        start=start_value,
        end=end_value,
        interpolate_fn=INTERPOLATES["interpolate"],
        easing_fn=EASINGS["ease_in_out_elastic"],
        amplitude=amplitude,
        period=period
    )
    y = np.array([fn(float(x)) for x in u])
    ax.plot(u, y, linewidth=2, label=f'amp={amplitude}, period={period}')
ax.plot([0, 1], [0, 1], '--', color='gray', alpha=0.6)
ax.set_title('make_ease_in_out_elastic(amplitude, period)')
ax.set_xlim(0, 1)
ax.grid(alpha=0.25)
ax.legend()

plt.tight_layout()
plt.show()
```
